```python
import pandas as pd
df= pd.read_parquet("/Users/anusha/CiteConnect/data/AI_in_healthcare_1761089776.parquet")
successful = df[df['status'].str.startswith('success')]

for idx, row in successful.iterrows():
    intro = row['introduction']
    
    print(f"\n{'='*60}")
    print(f"{row['title'][:60]}...")
    print(f"Length: {len(intro)} chars")
    print(f"\nFIRST 200 chars:\n{intro[:200]}")
    print(f"\nLAST 200 chars:\n{intro[-200:]}")
    print(f"\n{'='*60}")
    
    # Check if it bleeds into next section
    if '2.' in intro[-500:] or 'Method' in intro[-500:] or 'Materials and Methods' in intro[-500:]:
        print("‚ö†Ô∏è  WARNING: Might include next section!")
```

```python

# Set display options to see full content
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)

print("="*60)
print("SUCCESSFUL EXTRACTIONS")
print("="*60)

# Filter for successful ones
successful = df[df['status'].str.startswith('success')]

print(f"\nFound {len(successful)} successful extractions\n")

for idx, row in successful.iterrows():
    print(f"\n{'='*60}")
    print(f"Title: {row['title']}")
    print(f"Year: {row['year']}")
    print(f"Status: {row['status']}")
    print(f"URL: {row['pdf_url']}")
    print(f"{'='*60}")
    print(f"\nABSTRACT:")
    print(row['abstract'][:500] if row['abstract'] else "No abstract")
    print(f"\nINTRODUCTION:")
    print(row['introduction'][:1000] if row['introduction'] else "No introduction")
    print(f"\n... (truncated at 1000 chars)")
    print(f"\nFull intro length: {len(row['introduction']) if row['introduction'] else 0} characters")

print("\n" + "="*60)
print("ALL PAPERS SUMMARY")
print("="*60)
print(df[['title', 'status', 'pdf_url']].to_string())
```
